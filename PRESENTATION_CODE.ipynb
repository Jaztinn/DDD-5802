```python
import numpy as np
import matplotlib.pyplot as plt

def simple_convolution(signal, kernel):
    """Perform convolution of a signal with a kernel."""
    output_len = len(signal) + len(kernel) - 1
    output = np.zeros(output_len)

    for i in range(output_len):
        for k in range(len(kernel)):
            if 0 <= i - k < len(signal):
                output[i] += signal[i - k] * kernel[k]

    return output

signal = [1, 2, 1]
kernel = [1, 0.5]
output = simple_convolution(signal, kernel)


plt.figure(figsize=(10, 6))
plt.plot(signal, marker='o', label='Input Signal')
plt.title('Input Signal')
plt.grid()
plt.legend()

plt.figure(figsize=(10, 6))
plt.plot(kernel, marker='o', label='Kernel', color='orange')
plt.title('Kernel')
plt.grid()
plt.legend()

plt.figure(figsize=(10, 6))
plt.plot(output, marker='o', label='Convolution Output', color='green')
plt.title('Convolution Output')
plt.grid()
plt.legend()

plt.tight_layout()
plt.show()
```
